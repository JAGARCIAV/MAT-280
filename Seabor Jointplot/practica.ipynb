```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
iris
```

```python
sns.set_style("dark")
g=sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
plt.show()
```

```python
sns.set_style("dark")

g=sns.jointplot(x="sepal_length", y="sepal_width",
                data=iris, s=100,
                marginal_ticks=True,
                marginal_kws=dict(bins=25, fill=18))
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Longitud", "Sepal Anchura", fontsize=18)
plt.show()
```

```python
sns.set_style("dark")
g=sns.jointplot(x="sepal_length", y="sepal_width",
                data=iris, kind="reg",
                marginal_ticks=True,
                marginal_kws=dict(bins=45, fill=False))
g.fig.set_size_inches(10,10)
g.set_axis_labels("Sepal Longitud", "Sepal Anchura", fontsize=18)
plt.show()
```

```python
sns.set_style("dark")
g=sns.jointplot(x="sepal_length", y="sepal_width",
                data=iris, s=100,
                marginal_ticks=True)

g.plot_joint(sns.kdeplot, color="grey")
g.fig.set_size_inches(8,8)
g.set_axis_labels("Sepal Longitud", "Sepal Anchura", fontsize=18)
plt.show()
```
