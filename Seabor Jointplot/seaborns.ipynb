```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="sepal_length", y ="sepal_width", data=iris)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x="sepal_length", y="sepal_width",
                        data=iris, ci=90)
plt.show()
```

```python
iris["species"].unique()
```

```python
iris["species1"]=iris['species'].replace({"setosa":1,"versicolor":2, "virginica":3})
iris
```

```python
plt.Figure(figsize=(8,6))
sns.regplot(x="species1", y="sepal_length",
            data=iris, ci=90)
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.regplot(x="species1", y="sepal_length",
            data=iris, ci=90, x_jitter=0.04)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="species1", y="sepal_length",
            data=iris, ci=90, x_jitter=0.04)
plt.xticks([1,2,3], iris["species"].unique())
plt.xlabel("Species")
plt.ylabel("Sepal Lengeth")
plt.title("Sepal Lengrth for Species", fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="species1", y="sepal_length",
            data=iris, ci=90, x_estimator=np.median)
plt.xticks([1,2,3], iris["species"].unique())
plt.xlabel("Species")
plt.ylabel("Sepal Lengeth")
plt.title("Sepal Lengrth for Species", fontsize=18)
plt.xlim([0.8,3.2])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x="species1", y="sepal_length",
            data=iris, ci=90, x_estimator=np.mean)
plt.xticks([1,2,3], iris["species"].unique())
plt.xlabel("Species")
plt.ylabel("Sepal Lengeth")
plt.title("Sepal Lengrth for Species", fontsize=18)
plt.xlim([0.8,3.2])
plt.show()
```
