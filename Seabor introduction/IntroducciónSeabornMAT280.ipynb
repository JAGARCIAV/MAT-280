
# 2. Importando todas las bibliotecas requeridas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# 3. Importando la informacion

```python
mart=pd.read_excel("C:/Users/JOSE ARMANDO/Documents/tablas/Data/supermarket_sales_data.xlsx")
mart.head()
```

# 4. Crear un diagrama de conteo

```python
sns.countplot(x='Gender',data=mart)
```

```python
sns.countplot(x='Gender',data=mart, hue="Gender", palette="Set2")
plt.show()
```

# 5. Cree un (COUNTPLOT) básico para obtener el número de transacciones para cada línea de producto y cambie el tamaño de la figura

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line", data=mart, hue="Product line", palette="Set2")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Product line", data=mart, hue="Product line", palette="Set2", width=0.6)
plt.xticks(rotation=45)
plt.show()
```

# 6. Conviértalo en un gráfico de barras horizontales

```python
plt.figure(figsize=(15,5))
sns.countplot(y="Product line", data=mart, hue="Product line", palette="Set2")
plt.show()
```

# 7. Agregue tono para obtener el recuento en dos categorías, e.i. línea de producto y género.

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line", data=mart, hue="Gender", palette="Set2")
plt.show()
```

# 8. Usando diferentes paletas de colores

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line", data=mart, hue="Gender", palette="gist_gray")
plt.show()
```

# 9. Cambiar el estilo usando el color de la cara, el ancho de línea y el color del borde

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line', data=mart
            ,facecolor=(1,0,1,0)
            ,linewidth=5
            ,edgecolor=sns.color_palette('dark',3))
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line', data=mart
            ,facecolor=(1,0,1,0)
            ,linewidth=8
            ,edgecolor=sns.color_palette('dark',3))
```
