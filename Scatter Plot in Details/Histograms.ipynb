```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill")
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, y="total_bill")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, y="total_bill", binwidth=5)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill",  binwidth=5)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True,
             fill=False, element="step")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True,
             hue="time")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True,
             hue="time",
             multiple="stack")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True,
             hue="time",
             multiple="stack",
             kde=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", stat="probability")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True,
             fill=False, element="step", cumulative=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True,
             fill=False, element="step", cumulative=True, hue="day")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", log_scale=True,
             fill=False, element="step", cumulative=True, hue="day", hue_order=["Sat", "Sun"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="day", shrink=0.8, hue="smoker", multiple="dodge")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(tips, x="day", shrink=0.8, hue="smoker", multiple="stack")
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", y="tip")
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", y="tip", hue="sex")
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", y="day", hue="sex")
plt.show()
```

```python
plt.Figure(figsize=(8,6))
sns.histplot(tips, x="total_bill", y="day", hue="time")
plt.show()
```
