```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
ti=sns.load_dataset("titanic")
ti
```

```python
sns.set_theme(style="whitegrid")
```

```python
sns.pointplot(data=ti, x="class", y="age")
```

```python
plt.Figure(figsize=(8, 6))
sns.pointplot(data=ti, x="class", y="age")
```

```python
from numpy import median
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", estimator=median)
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", ci=90)
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", ci="sd")
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", hue="sex")
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", hue="sex", palette=["red", "green"])
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", hue="sex", palette=["red", "blue"],
                markers=["o", "s"], linestyles=["-", "--"])
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", hue="sex", palette=["red", "blue"],
                markers=["o", "s"], linestyles=["-", "--"],
                order=["First", "Third", "Second"])
```

```python
plt.Figure(figsize=(10, 6))
sns.pointplot(data=ti, x="class", y="age", capsize=0.05, join=False)
```

```python
figure = plt.Figure(figsize=(10, 6))
sns.stripplot(data=ti, x="class", y="age",alpha=0.5, palette="Set2")
sns.pointplot(data=ti, x="class", y="age", join=False, color="red")
```

```python
sns.boxplot(data=ti, x="class", y="age", palette="Set2")
sns.pointplot(data=ti, x="class", y="age",join=False,color="red")
```

```python
sns.violinplot(data=ti, x="class", y="age", palette="Set2")
sns.pointplot(data=ti, x="class", y="age",join=False,color="red")
```

```python
sns.catplot(data=ti, x="class", y="age", kind="point", hue="alive", col="embark_town", height=4, aspect=0.7)
```
